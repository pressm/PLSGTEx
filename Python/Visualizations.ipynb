```python
fig, ax = plt.subplots(1,3, figsize= (16,3), tight_layout=False)

ax[0].grid(False)
ax[0].set_facecolor('white')
ax[1].grid(False)
ax[1].set_facecolor('white')

data_shape = np.shape(d_corr_f)

# Take negative and positive data apart and cumulate
def get_cumulated_array(data, **kwargs):
    cum = data.clip(**kwargs)
    cum = np.cumsum(cum, axis=0)
    d = np.zeros(np.shape(data))
    d[1:] = cum[:-1]
    return d  

cumulated_data_f = get_cumulated_array(d_corr_f, lower=0)
cumulated_data_neg_f = get_cumulated_array(d_corr_f, upper=0)
cumulated_data_m = get_cumulated_array(d_corr_m, lower=0)
cumulated_data_neg_m = get_cumulated_array(d_corr_m, upper=0)

# Re-merge negative and positive data.
row_mask = (d_corr_f<0)
cumulated_data_f[row_mask] = cumulated_data_neg_f[row_mask]
data_stack_f = cumulated_data_f

row_mask = (d_corr_m<0)
cumulated_data_m[row_mask] = cumulated_data_neg_m[row_mask]
data_stack_m = cumulated_data_m

for i in np.arange(0, data_shape[0]):
    ax[0].bar(d_corr_f.columns, d_corr_f.values[i], bottom=data_stack_f[i], width = .5, color = sns.color_palette('tab10')[i])
for i in np.arange(0, data_shape[0]):
    ax[1].bar(d_corr_m.columns, d_corr_m.values[i], bottom=data_stack_m[i], width = .5, color = sns.color_palette('tab10')[i])


ax[0].set_title('Female')
ax[1].set_title('Male')
ax[0].set_ylabel('Correlation')
ax[1].set_ylabel('correlation')
ax[1].set_xlabel('Transcription Factors')

fig.legend(d_corr_f.index, loc = 'center left', ncol = 1)
ax[0].axhline(y=0, color = 'black')
ax[1].axhline(y=0, color = 'black')

sns.barplot(data = S, x = 'Channels', y = 'value', hue='Sex', ax = ax[2])
l0 = ax[0].get_xticklabels()
print(l0)
ax[2].grid(False)
ax[2].set_facecolor('white')
ax[2].axhline(y=0, color = 'black')
ax[2].set_ylabel('Sum of Correlation')
ax[2].set_xlabel('')
ax[0].set_xticklabels(d_corr_f.columns,rotation = 70)
ax[1].set_xticklabels(d_corr_m.columns,rotation = 70)
ax[2].set_xticklabels(ax[2].get_xticklabels(),rotation = 70)
```
